# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types
print(df.info())
print("\nMissing values per column:\n", df.isnull().sum())
```

```python

# Summary statistics
df.describe(include='all').transpose()

```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable

import matplotlib.pyplot as plt
target_col = 'y' if 'y' in df.columns else df.columns[-1]
target_counts = df[target_col].value_counts()
print("Target column:", target_col)
print(target_counts)

plt.figure(figsize=(6,4))
target_counts.plot(kind='bar')
plt.title('Distribution of target variable (' + target_col + ')')
plt.xlabel('Class')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python

# Correlation matrix for numerical features
import seaborn as sns
import matplotlib.pyplot as plt

num_df = df.select_dtypes(include=['int64','float64'])
corr = num_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", linewidths=0.5)
plt.title('Correlation matrix (numerical features)')
plt.tight_layout()
plt.show()

```
